```python
from sklearn.preprocessing import StandardScaler
df1 = df[['Amount', 'Time']]
cols = df1.columns
ss = StandardScaler()
df1 = ss.fit_transform(df1)
df1 = pd.DataFrame(df1, columns = cols)

df2 = df.drop(columns = ['Amount', 'Time'])
df2 = pd.concat([df2, df1], axis = 1)
df2
```

```python
cor = df2.corr()
plt.figure(figsize = (30,20))
ax = sns.heatmap(cor, annot = True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

```python
X = df.drop(['Class'], axis = 1)
y = df['Class']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_pred))

print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score,roc_curve

def model_compare(lr,X_train,X_test,y_train,y_test):
    lr.fit(X_train,y_train)
    ypred_train = lr.predict(X_train)
    ypred_test = lr.predict(X_test)
    
    yprob_train = lr.predict_proba(X_train)[:,1]
    yprob_test = lr.predict_proba(X_test)[:,1]
    
    print('Accuracy on Train Set',accuracy_score(y_train,ypred_train))
    print('Accuracy on Test Set',accuracy_score(y_test,ypred_test))
    
    print('AUC ROC on Train Set',roc_auc_score(y_train,yprob_train))
    print('AUC ROC on Test Set',roc_auc_score(y_test,yprob_test))
    
    fpr,tpr,thresholds = roc_curve(y_test,yprob_test)
    plt.plot(fpr,tpr)
    plt.plot(fpr,fpr,'r-')
    plt.show()
```

```python
model_compare(lr, X_train, X_test, y_train, y_test)
```

## Undersample the Majority Class

```python
print(len(y_train), len(y_test))

print(len(y_train[y_train == 1]), len(y_test[y_test == 1]))
```

```python
Xy_train = pd.concat([X_train, y_train], axis = 1)

Xy_train0 = Xy_train.loc[Xy_train['Class'] == 0]
Xy_train1 = Xy_train.loc[Xy_train['Class'] == 1]

len0 = len(Xy_train0)
len1 = len(Xy_train1)

Xy_train0_us = Xy_train0.sample(len1)

Xy_train_us = pd.concat([Xy_train0_us, Xy_train1], axis = 0)

X_train_us = Xy_train_us.drop(['Class'], axis = 1)
y_train_us = Xy_train_us['Class']
```

We undersample the majority class i.e. 0 to make equal to minority class i.e 1. This excersise was done only for train dataset Test data was untouch

```python
lr1 = LogisticRegression()
model_compare(lr1, X_train_us, X_test, y_train_us, y_test)
```

## Oversampling

```python
Xy_train = pd.concat([X_train, y_train], axis = 1)

Xy_train0 = Xy_train.loc[Xy_train['Class'] == 0]
Xy_train1 = Xy_train.loc[Xy_train['Class'] == 1]

len0 = len(Xy_train0)
len1 = len(Xy_train1)

Xy_train1_os = Xy_train1.sample(len0, replace = True)

Xy_train_os = pd.concat([Xy_train1_os, Xy_train0], axis = 0)

X_train_os = Xy_train_os.drop(['Class'], axis = 1)
y_train_os = Xy_train_os['Class']
```

```python
lr2 = LogisticRegression()
model_compare(lr2, X_train_os, X_test, y_train_os, y_test)
```

## Smoting

Smoting is nothing but the oversampling but not blindly multiplying to one's just creating the new data points

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy = 'minority')
X_train_sm, y_train_sm = smote.fit_sample(X_train, y_train)
```

```python
lr3 = LogisticRegression()
model_compare(lr3, X_train_sm, X_test, y_train_sm, y_test)
```
