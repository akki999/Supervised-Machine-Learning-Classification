## Logistic Regression - BY Statistical Approach

```python
import statsmodels.api as sm
```

```python
y = df['TenYearCHD']
X = df.drop('TenYearCHD', axis = 1)
```

```python
X = sm.add_constant(X)
```

```python
model = sm.Logit(y, X)
result = model.fit()
result.summary()
```

```python
## Calculating VIF for all the variables

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
pd.DataFrame({'vif': vif}, index=X.columns).T
```

```python
## Backward elimination to drop insignificant variables one by one
     
```

```python
cols = list(X.columns)
p = []

while len(cols) > 1:
    X = X[cols]
    model = sm.Logit(y,X).fit().pvalues
    p = pd.Series(model.values[1:], index = X.columns[1:])
    pmax = max(p)
    pid = p.idxmax()
    if pmax > 0.05:
        cols.remove(pid)
        print('Variable Removed:', pid, pmax)
    else:
        break
cols
```

```python
model1 = sm.Logit(y, X[cols])
result1 = model1.fit()
result1.summary()
```

For Age Three conclusions
1)By looking sign of coeff of age we can say that as age increases CHD increases
2)As age increases by 1 log(odds) of CHD increases by 0.068.
3)As age increases by 1 odds of CHD increases by 7%.

For Male:
1)By looking the sign of coeff of male we can say that compare to female male got higher chance of CHD.
2)If it is male log(odds) of CHD for male is 0.4770 higher than male
3)Male got 61% chance of getting CHD than female.

For cigsperday:
1)By looking sign of coeff of Cigsperday we can say that as Cigsperday increases CHD increases.
2)As Cigsperday increases by 1 log(odds) of CHD increases by 0.0221.
3)As Cigsperday increases by 1 odds of CHD increases by 2.2%

## ROC Analysis

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(solver = 'liblinear', fit_intercept = True)

logreg.fit(X_train, y_train)

y_prob_train = logreg.predict_proba(X_train)[:,1] 
y_pred_train = logreg.predict(X_train)

y_prob_test = logreg.predict_proba(X_test)[:,1] 
y_pred_test = logreg.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

print('AUC of Train:', roc_auc_score(y_train, y_prob_train))

fpr, tpr, thresholds = roc_curve(y_train, y_prob_train)


plt.plot(fpr, tpr)
plt.plot(fpr,fpr, 'r-')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```

```python

print('AUC of Train:', roc_auc_score(y_test, y_prob_test))

fpr, tpr, thresholds = roc_curve(y_test, y_prob_test)


plt.plot(fpr, tpr)
plt.plot(fpr,fpr, 'r-')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC for Test Data')
plt.show()
```

```python
thresholds[thresholds > 1] = 1
```

```python
thresholds
```

```python
print('AUC of Train:', roc_auc_score(y_train, y_prob_train))
plt.plot(fpr, tpr)
plt.plot(fpr,fpr, 'r-')
plt.plot(fpr, thresholds)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```

```python

print('AUC of train: ', roc_auc_score(y_train, y_prob_train)) #for train data

fig, ax = plt.subplots()

ax.plot(fpr, tpr) #blue color
ax.plot(fpr, fpr, 'r-') #red color
plt.xlabel('FPR')
plt.ylabel('TPR')
ax1 = ax.twinx()
ax1.plot(fpr,thresholds) #orange color 


plt.title('ROC_Curve of Train data')
plt.show()
```
